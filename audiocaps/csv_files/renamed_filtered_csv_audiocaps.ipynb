```python
def load_picke_file(file_name):

    with open(file_name, 'rb') as f:
        pickle_obj = pickle.load(f)
    return pickle_obj
```

```python
def _sentence_process(sentence, add_specials=False):

    # transform to lower case
    sentence = sentence.lower()

    if add_specials:
        sentence = '<sos> {} <eos>'.format(sentence)

    # remove any forgotten space before punctuation and double space
    sentence = sub(r'\s([,.!?;:"](?:\s|$))', r'\1', sentence).replace('  ', ' ')

    # remove punctuations
    sentence = sub('[,.!?;:\"]', ' ', sentence).replace('  ', ' ')

    return sentence
```

```python
vocabulary = load_picke_file(vocabulary_path)
```

```python

keys_cap = list(captions.keys())
for key in keys_cap:
    sentences = captions[key]
    ok = 0
    for sentence in sentences:
        sentence_combined = ' '.join(sentence)
        sentence_processed = _sentence_process(sentence_combined)
        words = sentence_processed.strip().split()
        for word in words:
            try:
                vocabulary.index(word)
            except ValueError:
                ok = 1
                break
        if ok == 1:
            break
    if ok == 1:
        print(key)
```

```python
def txt_to_csv(dataset, split, captions, dataset_splits):
    with open(f"/scratch/shared/beegfs/oncescu/coding/libs/pt/DCASE2021_task6_v2/{dataset}/csv_files/vgg_{split}.csv", "w", newline='') as f:
        w = csv.writer(f)
        w.writerow(["file_name", "caption"])
        for element in dataset_splits[split]:
            sentences = captions[element]
            for sentence in sentences:
                sentence_caption = ' '.join(sentence)
                if sentence_caption[0] == '"':
                    sentence_caption = sentence_caption[1:-1]
                w.writerow([f"{element}.wav", sentence_caption])
```
